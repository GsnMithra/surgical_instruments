```python
import os
import math

path_train = 'surgical_dataset/images/train'
path_val = 'surgical_dataset/images/val'

train_size = len (os.listdir (path_train))
val_size = len (os.listdir (path_val))

print (f'train: {train_size}')
print (f'val: {val_size}')
print (f'total: {train_size + val_size}')

batch = int (input ('batch size: '))
epochs = int (input ('epochs: '))

def roundTo32(x, base = 32):
    return int(base * math.floor(float(x) / base))

img = roundTo32 ((train_size + val_size) / batch)

print (f'\nimg: {img}\nbatch:{batch}\nepochs:{epochs}')
```

```python
!python3 yolov5/train.py --img {img} --batch {batch} --epochs {epochs} --data yolov5/data/surgical_data.yaml --weights yolov5s.pt --cache
```
